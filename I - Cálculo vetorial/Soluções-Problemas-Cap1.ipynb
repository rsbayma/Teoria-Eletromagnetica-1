```python
# (1.5)
a, b = sp.symbols(['alpha', 'beta'])
A = sp.Matrix([5,3,2])
B = sp.Matrix([-1,4,6])
C = sp.Matrix([8,2,0])
v = a*A+b*B+C
v
```

```python
from IPython.display import display, Math
sist = [sp.Eq(v[0],0),sp.Eq(v[2],0)]
sol = sp.solve(sist)
display(Math(sp.latex(sist)))
display(Math(sp.latex(sol)))
```

```python
v.subs(sol)
```
